```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from google.colab import drive
drive.mount('/content/drive')
p1 = '/content/drive/MyDrive/Colab Notebooks/train.csv'
data = pd.read_csv(p1)
```

```python
data.head()
```

```python
s=round((data["Sex"].value_counts())/len(data)*100,2)
pd.DataFrame(s)
```

```python
labels = ['male','female']

sizes = data.Sex.value_counts()

figi,axi = plt.subplots()

axi.pie(sizes,labels=labels, autopct="%1.1f%%", shadow=True, colors = ['Blue' ,'Orange'])
plt.show()
```

```python
plt.figure()
cat= data[data.Sex=='male'].plot.scatter('Age',"Fare",color='blue',label='Male')
data[data.Sex=='female'].plot.scatter('Age',"Fare",color='yellow',label='Female',ax=cat)
```
